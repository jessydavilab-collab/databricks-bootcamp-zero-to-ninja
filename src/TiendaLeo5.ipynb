```python
#desc_nulls_first
from pyspark.sql import Row
df_desc = spark.createDataFrame([('Tom',80),(None,60),('Alice',None),(None,None)],["name","height"])
df_desc.select(df_desc.name).orderBy(df_desc.name.desc_nulls_first()).collect()
```

```python

#dropFields()
from pyspark.sql import Row
from pyspark.sql.functions import col, lit

df_drop= spark.createDataFrame([Row(a=Row(b=1,c=2, d=3, e=Row(f=4,g=5,h=6)))])
df_drop.show()
df_drop.withColumn('a',df_drop['a'].dropFields('b')).show()
```

```python
#dropFields()
from pyspark.sql import Row
df_drop2= spark.createDataFrame([Row(a=Row(b=1,c=2, d=3, e=Row(f=4,g=5,h=6)))])
df_drop2.withColumn('a',df_drop2['a'].dropFields('b','c')).show()
```

```python
#dropFields()
from  pyspark.sql import Row
df_drop3 = spark.createDataFrame([Row(a=Row(b=1,c=2,d=3, e=Row(f=4,g=5,h=6)))])
df_drop3.withColumn("a",col("a").dropFields("e.g","e.h")).show()
```

```python
#dropFields()
from pyspark.sql import Row
from pyspark.sql.functions import col, lit
df_drop4 = spark.createDataFrame([Row(a=Row(b=1,c=2,d=3, e=Row(f=4, g=5, h=6)))])
df_drop4.select(col("a").withField("e", col("a.e").dropFields("g","h")).alias("i")).show()
```

```python
#endswith
df_end = spark.createDataFrame([(2, 'Alice'),(5,'Bob'), (None,"Jessica"),(None,'Maria'),(58,None)],["age","name"])
df_end.filter(df_end.name.endswith('a')).collect()
```

```python
#endswith + $
df_end2 = spark.createDataFrame([(2, 'Alice'),(5,'Bob'), (None,"Jessica"),(None,'Maria'),(58,None)],["age","name"])
df_end2.filter(df_end2.name.endswith('ice$')).show()
```

```python
#startswith
df_start = spark.createDataFrame([(2, 'Alice'),(5,"Bob"),(None, None),(None,"Maria"),(67,None)],["age","name"])

df_start.filter(df_start.name.startswith('Ma')).show()
```

```python
from pyspark.sql import Row
df_eqNull = spark.createDataFrame([Row(id=1, value='foo'),Row(id=2, value=None),Row(id=3,value='food') ])
df_eqNull.select(df_eqNull['value']=='foo',
                 df_eqNull['value'].eqNullSafe('foo'),
                 df_eqNull['value'].eqNullSafe(None)
                 ).show()
```

```python
#getField()
from pyspark.sql import Row
df_getF = spark.createDataFrame([Row(r=Row(a=1,b="b"),c=Row(d=2,e=3,f="f"))])
#df_getF.show()
df_getF.select(df_getF.c.getField("f")).show()
```

```python
#getItem()
df_getI = spark.createDataFrame([([1,2,3],{"Key": "value"})], ["l","d"])
df_getI.select(df_getI.l.getItem(1), df_getI.d.getItem("Key")).show()
```

```python
#ilike
df_ili = spark.createDataFrame([(2,"Alice"),(5,"Bob"),(3,"Ailen"),(4,"Marien")], ["age","name"])
df_ili.filter(df_ili.name.ilike('%en')).show()
```
