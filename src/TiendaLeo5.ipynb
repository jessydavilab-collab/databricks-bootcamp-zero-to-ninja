```python
#desc_nulls_first
from pyspark.sql import Row
df_desc = spark.createDataFrame([('Tom',80),(None,60),('Alice',None),(None,None)],["name","height"])
df_desc.select(df_desc.name).orderBy(df_desc.name.desc_nulls_first()).collect()
```

```python

#dropFields()
from pyspark.sql import Row
from pyspark.sql.functions import col, lit

df_drop= spark.createDataFrame([Row(a=Row(b=1,c=2, d=3, e=Row(f=4,g=5,h=6)))])
df_drop.show()
df_drop.withColumn('a',df_drop['a'].dropFields('b')).show()
```

```python
#dropFields()
from pyspark.sql import Row
df_drop2= spark.createDataFrame([Row(a=Row(b=1,c=2, d=3, e=Row(f=4,g=5,h=6)))])
df_drop2.withColumn('a',df_drop2['a'].dropFields('b','c')).show()
```

```python
#dropFields()
from  pyspark.sql import Row
df_drop3 = spark.createDataFrame([Row(a=Row(b=1,c=2,d=3, e=Row(f=4,g=5,h=6)))])
df_drop3.withColumn("a",col("a").dropFields("e.g","e.h")).show()
```

```python
#dropFields()
from pyspark.sql import Row
from pyspark.sql.functions import col, lit
df_drop4 = spark.createDataFrame([Row(a=Row(b=1,c=2,d=3, e=Row(f=4, g=5, h=6)))])
df_drop4.select(col("a").withField("e", col("a.e").dropFields("g","h")).alias("i")).show()
```
