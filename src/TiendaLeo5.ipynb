```python
#otherwise()
from pyspark.sql import functions as sf 
df_other = spark.createDataFrame([(2,"Alice"),(4,"Bob"),(3,"gato")],["age","name"])
df_other.select(df_other.name, sf.when(df_other.age > 3,1).otherwise(0)).show()
```

```python
#over()
from pyspark.sql import Window
from pyspark.sql.functions import rank, min, desc
window = (
    Window.partitionBy("name")
    .orderBy("age")
    .rowsBetween(Window.unboundedPreceding, Window.currentRow)
)
df_over = spark.createDataFrame(
    [(2, "Alice"),(5,"Bob"),(7,"Jessy"),(7,"Jessy")],["age","name"])
df_over.withColumn("rank", rank().over(window)
    ).withColumn("nim", min('age').over(window)
    ).sort(desc("age")).show()
```

```python
#when()
from pyspark.sql import functions as sf
df_when = spark.createDataFrame([(2,"Alice"), (5,"Bob"),(4,"Jessy")], ["age","name"])
df_resul = df_when.select(df_when.name, sf.when(df_when.age > 4, 1 ).when(df_when.age < 3, -1).otherwise(0))
df_resul.show()
```

```python
#when()
from pyspark.sql import functions as sf
df_when2 = spark.createDataFrame([(1,"Alice"),(2,"Bob"),(6,"Charlie"),(4,"Jessy")],("age","name"))
df_resul2=df_when2.select(
    df_when2.name, sf.when(df_when2.age < 3, "Young").when(df_when2.age < 5, "Middle-aged").otherwise("Old")
)
df_resul2.show()
```

```python
from pyspark.sql import functions as sf
df_when3 = spark.createDataFrame([(2,"Alice"),(5,"Bob"),(0,"jessy")],["age","name"])
df_result3 = df_when3.select(df_when3.name,
                            sf.when(sf.lit(True), 1).otherwise(
                            sf.raise_error("unreachable")).alias("when"))
df_result3.show()
```
