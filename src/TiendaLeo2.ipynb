```python
from pyspark.sql import *


df_children = spark.createDataFrame(
    data = [("Jesus", 15), ("Zaky", 13),("Zoya", 8),("Renzo", 48), ("Platanitos", 9)],
    schema = ["name string","age int"])
display(df_children)
```

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

df_children_with_schema = spark.createDataFrame(
   data= [("Mikhail", 15), ("Zaky", 13), ("Zoya", 8)],
   schema = StructType ([
       StructField('name', StringType(), True),
       StructField('age', IntegerType(), True) 
       
   ])
) 
display(df_children_with_schema)
```

```python
df_customer = spark.table('samples.tpch.customer')
display(df_customer)
```

```python
from pyspark.sql.functions import col
df_customer.select(
    col("c_custkey"),
    col("c_acctbal")
).show()


```

```python
from pyspark.sql.functions import expr
display(df_customer.select(
    expr("c_custkey"),
    expr("c_acctbal")

) )
```

```python
#from pyspark.sql.functions import selectExpr
display( df_customer.selectExpr(
    "c_custkey as key",
    "round(c_acctbal) as account_rounded"
))
```

```python
df_customer.select(
    "c_custkey",
    "c_acctbal"
).show()
```

```python
df_customer.select(
    df_customer.c_custkey.alias("key") ,
    df_customer.c_acctbal.alias("numeros")
).show()
```

```python
display(df_customer.select(
    df_customer["c_custkey"].alias("key2"),
    df_customer["c_acctbal"].alias("decimal")
) )
```

```python
from pyspark.sql.functions import col
df_customer_flag = df_customer.withColumn("balance_flag", col("c_acctbal") < 500)
display(df_customer_flag) c_phone, c_mktsegment
```

```python
df_customer_flag_renamed = df_customer_flag.withColumnRenamed("balance_flag","balance_flag_renamed")
df_customer_flag_renamed.show()
```

```python
df_customer_flag_renamed = df_customer_flag.withColumnRenamed("balance_flag","balance_flag_renamed")
df_customer_flag_renamed_filtro = df_customer_flag_renamed.select("balance_flag_renamed")
display(df_customer_flag_renamed_filtro)
```

```python
from pyspark.sql.functions import avg

df_segment_balance = df_customer.groupBy("c_mktsegment").agg(avg(df_customer["c_acctbal"]).alias("avg_account_balance"))
display(df_segment_balance) 
```
