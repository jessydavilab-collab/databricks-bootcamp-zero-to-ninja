**Anexar Filas con otros data frame**

```python
#Anexar Fila
from pyspark.sql.functions import col
df_that_one = df.filter(col("Count") == 29)
display(df_that_one)
 
```

```python
#Anexar Fila
df_filtered = df.filter((col("Count") == 40) | (col("Count") == 44))
display(df_filtered)
```

```python
#Anexar Fila
df_anexar_fila = df_that_one.union(df_filtered)

display(df_anexar_fila)
```
