```python
df_practice = spark.createDataFrame([(2,"Matrin"),(5,"Leo"),(7,"Chappie"),(9,"Hidalgo"),(7,"Maria"),(10,"Ana"),(7,"Arturo"),(None,None),(6,None),(None,"David"),(55,"Blakie")],["age","name"])
df_practice.show()
```

```python
# Arithmetic operations
df1 = df_practice
df1.select((df1.age + 10).alias("en 10 años")).show()
```

```python
# Comparison operations
df2 = df_practice
df2.filter(df2.age >3).show()
```

```python
# String operations
df3 = df_practice
df3.filter(df3.name.startswith("A")).show()
```

```python
# Null checking
df4 = df_practice
df4.filter(df4.name.isNotNull() | df4.age.isNotNull()).show()
```

```python
from pyspark.sql import functions as F

df5 = df_practice
df5.select( df5.name, df5.age,
           F.when(df5.age < 3, "Child")
            .when(df5.age < 13, "Kid")
            .otherwise("adult")
            .alias("age_group")
).show()
```

```python
df6 = df_practice
df6.orderBy(df6.age.desc()).show()
```

```python
df7 = df_practice
df7.orderBy(df7.age.asc_nulls_last()).show()
```

```python
#agg()
from pyspark.sql import functions as sf
#df_agg = spark.createDataFrame([(2, "Alice"),(5,"Bob")],["age","name"])
df_agg = df_practice
df_agg.agg({"age": "max"}).show()

df_agg.agg(sf.min(df_agg.age)).show()
```
